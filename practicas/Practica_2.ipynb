# Tarea 3: Ingreso y patrones de consumo en México 

## Contexto

El análisis del consumo de los hogares es clave para entender cómo funciona la economía en la vida cotidiana. En esta práctica se utilizará una selección de variables de la Encuesta Nacional de Ingreso y Gasto en los Hogares (ENIGH), elaborada por el INEGI. A partir de datos reales, se analizará la relación entre el ingreso corriente del hogar y distintos tipos de gasto, como alimentos, vivienda, transporte, educación, salud y esparcimiento.

El objetivo es identificar qué gastos tienden a aumentar conforme crece el ingreso, cuáles se mantienen relativamente constantes y cuáles están asociados a estilos de vida de mayores niveles socioeconómicos. El énfasis estará en el uso de estadística descriptiva, visualización y correlaciones como herramientas exploratorias, reconociendo tanto su utilidad como sus limitaciones.

La liga al archivo es: https://github.com/adan-rs/amd/raw/main/data/enigh2025.xlsx

## Objetivos de aprendizaje

Al finalizar esta práctica, el estudiante será capaz de:
1. Analizar datos reales de hogares mexicanos, interpretando distribuciones asimétricas.
2. Interpretar estadísticos descriptivos y percentiles para describir la heterogeneidad del ingreso y el gasto.
3. Visualizar la relación entre ingreso y gasto e identificar patrones generales.
4. Utilizar correlaciones como herramienta exploratoria para analizar la relación del ingreso y distintos tipos de gasto.
5. Interpretar relaciones entre distintos rubros de gasto desde una perspectiva económica.

*Hint: no se espera que descubras "verdades absolutas", se espera que uses los datos para argumentar reconociendo las limitaciones del análisis*
